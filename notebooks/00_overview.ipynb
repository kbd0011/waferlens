# 00 — Overview & synthetic smoke run

Run `make demo` first, or generate synthetic data here and train a quick CNN.

```python
from waferlens.data.synthetic import generate, BASE_CLASSES
import matplotlib.pyplot as plt
X, Y = generate(n=12, size=52, seed=0)
fig, axes = plt.subplots(2, 6, figsize=(14, 5))
for i, ax in enumerate(axes.flat):
    ax.imshow(X[i], cmap='viridis'); ax.axis('off')
    labels = [BASE_CLASSES[j] for j in range(8) if Y[i, j] > 0]
    ax.set_title(', '.join(labels) or 'none', fontsize=8)
plt.tight_layout(); plt.show()
```
