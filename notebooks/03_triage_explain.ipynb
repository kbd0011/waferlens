# 03 — Triage + explainability on a single wafer

```python
from waferlens.config import load_config
from waferlens.data.mixedwm38 import load_npz
from waferlens.predict import load_checkpoint, predict_map
import matplotlib.pyplot as plt

cfg = load_config('configs/config.yaml')
d = load_npz('data/synthetic_sample.npz')
net, classes, thr = load_checkpoint('artifacts/checkpoints/cnn_best.pt', cfg.model, 52)
wmap = d.X[7]
pred = predict_map(net, classes, wmap, threshold=thr, with_saliency=True)
print('detected:', pred.detected)
for e in pred.triage['entries']:
    print(e['pattern'], '->', e['candidate_causes'][0])

fig, ax = plt.subplots(1, 2, figsize=(9, 4))
ax[0].imshow(wmap, cmap='viridis'); ax[0].set_title('wafer'); ax[0].axis('off')
if pred.saliency is not None:
    ax[1].imshow(wmap, cmap='gray'); ax[1].imshow(pred.saliency, cmap='jet', alpha=0.5)
    ax[1].set_title('saliency'); ax[1].axis('off')
plt.tight_layout(); plt.show()
```
